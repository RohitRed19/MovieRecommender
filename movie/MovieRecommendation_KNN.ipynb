```python
import pandas as pd
from scipy.sparse import csr_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
from fuzzywuzzy import process
```

```python
movies='movies.csv'
rating='ratings.csv'

df_movies=pd.read_csv(movies, usecols=['movieId','title'],dtype={'movieId':'int32','title':'str'})
df_rating=pd.read_csv(rating, usecols=['userId', 'movieId','rating'],dtype={'userId':'int32','movieId':'int32','rating':'float'})
```

```python
df_movies
```

```python
#sparse matrix
movies_users=df_rating.pivot(index='movieId',columns='userId',values='rating').fillna(0)
mat_movies_users=csr_matrix(movies_users.values)
```

```python
movies_users
```

```python
#cosine similarity
model_knn= NearestNeighbors(metric='cosine', algorithm='brute',n_neighbors=20)
```

```python
#Recommender(movie_name)=> List of Movie recommended

def recommender(movie_name,data,model,n_recommendations):
    model.fit(data)
    idx=process.extractOne(movie_name, df_movies['title'])[2]
    print('Movie Selected:',df_movies['title'][idx],'Index: ',idx)
    print('Searching for recommendations....')
    distances, indices=model.kneighbors(data[idx], n_neighbors=n_recommendations)
    for i in indices:
        print(df_movies['title'][i].where(i!=idx))
        
#         
    
```

```python
recommender('Neeraj',mat_movies_users,model_knn,20)
```
